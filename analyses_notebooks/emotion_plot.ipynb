```python
import pandas as pd
```

```python
dataset = 'datasets/raw/emotions_generated.csv'
```

```python
emotions = ['Anger', 'Awe', 'Contentment', 'Disgust', 'Excitement', 'Fear', 'Sadness', 'Something Else']
```

```python
def select_emotion(row):
    index = row[emotions].values.argmax()
    return emotions[index]
```

```python
df = pd.read_csv(dataset)
```

```python
df['selected'] = df.apply(select_emotion, axis=1)
```

```python
df_stylegan2 = df[df['type'] == 'StyleGAN2']
df_stylecan2 = df[df['type'] == 'StyleCAN2']
```

```python
df_stylegan2['selected'].value_counts().plot(kind='bar')
```
